```python
xmin = 0
ymin = 0
zmin = 0
xmax = 10
ymax = 10
zmax = 10
ncell = 10

dx = (xmax-xmin)/ncell
dy = (ymax-ymin)/ncell
dz = (zmax-zmin)/ncell

xprimal = np.arange(xmin,xmax+dx,dx)
yprimal = np.arange(ymin,ymax+dy,dy)
zprimal = np.arange(zmin,zmax+dz,dz)

xdual = np.arange(xmin-dx/2,xmax+3*dx/2,dx)
ydual = np.arange(ymin-dy/2,ymax+3*dy/2,dy)
zdual = np.arange(zmin-dx/2,zmax+3*dz/2,dz)

print("xp=", xprimal, '\n')
print("xd=", xdual, '\n')
poldeg = 2

c0x = xprimal**poldeg
c0x[0] = (c0x[0]+c0x[-1])/2 # periodicity 
c0x[-1] = c0x[0]
c0y = yprimal
c0y[0] = (c0y[0]+c0y[-1])/2 # periodicity 
c0y[-1] = c0y[0]
c0z = zprimal
c0z[0] = (c0z[0]+c0z[-1])/2 # periodicity 
c0z[-1] = c0z[0]

c1x = 1/(poldeg+1)*(xprimal[1:]**(poldeg+1)-xprimal[:-1]**(poldeg+1))
c1y = dy * xprimal**poldeg
c1y[0] = (c1y[0]+c1y[-1])/2 # periodicity 
c1y[-1] = c1y[0]
c1z = dz * xprimal**poldeg


c2x = dy * dz * xprimal**poldeg
c2x[0] = (c2x[0]+c2x[-1])/2 # periodicity 
c2x[-1] = c2x[0]
c2y = 1/(poldeg+1)*(xprimal[1:]**(poldeg+1)-xprimal[:-1]**(poldeg+1)) * dz
c2y[0] = (c2y[0]+c2y[-1])/2 # periodicity 
c2y[-1] = c2y[0]
c2z = 1/(poldeg+1)*(xprimal[1:]**(poldeg+1)-xprimal[:-1]**(poldeg+1)) * dy



#print(len(c0x), "c0x = ",c0x, '\n')
print(len(c1x), "Ex = ",c1x, '\n')
print(len(c1y), "Ey = ",c1y, '\n')
print(len(c2x), "Bx = ",c2x, '\n')
print(len(c2y), "By = ",c2y, '\n')

tc0x = xdual[1:-1]**poldeg

tc1x = 1/(poldeg+1)*(xdual[1:]**(poldeg+1)-xdual[:-1]**(poldeg+1))
tc1x[0] = (tc1x[0]+tc1x[-1])/2 # periodicity 
tc1x[-1] = tc1x[0]
tc1y = dy * xdual**poldeg
tc1y[0] = (tc1y[0]+tc1y[-1])/2 # periodicity 
tc1y[-1] = tc1y[0]

tc2x = dy * dz * xdual**poldeg
tc2y = 1/(poldeg+1)*(xdual[1:]**(poldeg+1)-xdual[:-1]**(poldeg+1))*dz
tc2y[0] = (tc2y[0]+tc2y[-1])/2 # periodicity 
tc2y[-1] = tc2y[0]

#print(len(tc0x),"tc0x=",tc0x, '\n')
print(len(tc1x),"Hx =",tc1x, '\n')
print(len(tc1y),"Hy =",tc1y, '\n')
print(len(tc2x),"Dx =",tc2x, '\n')
print(len(tc2y),"Dy =",tc2y, '\n')
```
