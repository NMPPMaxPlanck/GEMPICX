```python
import numpy as np
```

```python
xmin = 0
ymin = 0
zmin = 0
xmax = 10
ymax = 10
zmax = 10
ncell = 10
dx = (xmax-xmin)/ncell
dy = (ymax-ymin)/ncell
dz = (zmax-zmin)/ncell
xprimal = np.arange(xmin,xmax+dx,dx)
yprimal = np.arange(ymin,ymax+dy,dy)
zprimal = np.arange(zmin,zmax+dz,dz)
xdual = np.arange(xmin-dx/2,xmax+3*dx/2,dx)
ydual = np.arange(ymin-dy/2,ymax+3*dy/2,dy)
zdual = np.arange(zmin-dx/2,zmax+3*dz/2,dz)
print("xp=", xprimal)
print("xd=", xdual)
poldeg = 2

c0x = xprimal**poldeg
c0x[0] = (c0x[0]+c0x[-1])/2 # periodicity 
c0x[-1] = c0x[0]
c0y = yprimal**poldeg
c0y[0] = (c0y[0]+c0y[-1])/2 # periodicity 
c0y[-1] = c0y[0]

c1x = 1/(poldeg+1)*(xprimal[1:]**(poldeg+1)-xprimal[:-1]**(poldeg+1))
c1y = 1/(poldeg+1)*(yprimal[1:]**(poldeg+1)-yprimal[:-1]**(poldeg+1))
c1x = 1/(poldeg+1)*(zprimal[1:]**(poldeg+1)-zprimal[:-1]**(poldeg+1))


c2y = 1/(poldeg+1)*(xprimal[1:]**(poldeg+1)-xprimal[:-1]**(poldeg+1))*yprimal[1:]


print(len(c0x), "c0x=",c0x)
print(len(c1x), "c1x=",c1x)
print(len(c2y), "c2y=",c2y)

tc0x = xdual[1:-1]**poldeg

tc1x = 1/(poldeg+1)*(xdual[1:]**(poldeg+1)-xdual[:-1]**(poldeg+1))
tc1x[0] = (tc1x[0]+tc1x[-1])/2 # periodicity 
tc1x[-1] = tc1x[0]

tc2y = 1/(poldeg+1)*(xdual[1:]**(poldeg+1)-xdual[:-1]**(poldeg+1))*ydual[1:]
#tc2y[0] = (tc2y[0]+tc2y[-1])/2 # periodicity 
#tc2y[-1] = tc2y[0]

print(len(tc0x),"tc0x=",tc0x)
print(len(tc1x),"tc1x=",tc1x)
print(len(tc2y),"tc2y=",tc2y)
```
