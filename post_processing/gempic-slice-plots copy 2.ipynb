```python
# Create a line plot of the variables 'E_x', 'E_y' and 'B_z' with 100 sampling points evenly
# spaced between the coordinates (0, 1, 1) and (L[0], 1, 1)
plot = yt.LinePlot(ds, [('boxlib', 'rho')], (0, 1, 1), (L[0], 1, 1), 100)

# plot.save('./Plots/bernstein_initial_data.png')
plot.show()
```
