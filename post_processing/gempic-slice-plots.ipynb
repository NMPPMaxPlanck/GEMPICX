```python
# Import needed modules (needs yt>=4 version, do pip install 'yt>=4.0' if needed)
# Classical yt plot examples can be found in Cookbook https://yt-project.org/doc/cookbook
import os
import numpy as np
import matplotlib.pyplot as plt
import yt
from yt.frontends.boxlib.data_structures import AMReXDataset
```

```python
# Set working directory
pathname = '/Users/sonnen/Codes/gempic_runs/'
pathname_out = pathname + '/processed'
try:
    os.mkdir(pathname_out)
except(FileExistsError):
    pass
os.chdir(pathname)
sim_name = "Weibel"
```

```python
# Read data
slices =os.listdir("Plotfiles")
slices.sort()
print(slices)
iteration = 0
plotfile = './Plotfiles/{}{:05d}'.format(sim_name, iteration)
ds = AMReXDataset(plotfile)
print(ds.field_list)

step = int(slices[0][-5:])
time = float(ds.current_time)
x_left = np.array(ds.domain_left_edge)
x_right = np.array(ds.domain_right_edge)
L = x_right - x_left
print(time)
print(L[0],L[1],L[2])
```

```python
 # Slice plot of B_z (with and without saving to a file)
yt.SlicePlot(ds, 'y', "B_z").save()
yt.SlicePlot(ds, 'z', "B_z")
```

```python
# Create a line plot of the variables 'E_x', 'E_y' and 'B_z' with 100 sampling points evenly
# spaced between the coordinates (0, 1, 1) and (L[0], 1, 1)
yt.LinePlot(ds, [('boxlib', 'E_x'), ('boxlib', 'E_y'), ('boxlib', 'B_z')], (0, 1, 1), (L[0], 1, 1), 100)
```

```python
# particle plot (only for positions)
pxy=yt.ParticlePlot(
    ds,
    ("all", "particle_position_x"),
    ("all", "particle_position_y"),
    width=(0.5, 0.5)
)

pxy.show()
# particle plot colored by the particle weights
pxyw=yt.ParticlePlot(
    ds,
    ("all", "particle_position_x"),
    ("all", "particle_position_y"),
    [("all", "particle_weight")],
    width=(0.75, 0.75)  # particle size
)
pxyw.show()
```
