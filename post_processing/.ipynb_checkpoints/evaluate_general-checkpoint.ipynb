Loading data into numpy arrays:

```python
# get simulation name:
path = os.getcwd()
input_file = open(path + "/gempic_obj/gempic/simulations/vlasov_maxwell/" + input_name + '.input')
lines = input_file.readlines()
sim_name = lines[1]
# keep only what is between ""
sim_name = sim_name[sim_name.find('\"')+1:]
sim_name = sim_name[:sim_name.find('\"')]

# get file path
filename = path + "/gempic_obj/gempic/simulations/vlasov_maxwell/" + sim_name + ".tmp"

# read in data
n_lines = sum(1 for line in open(filename))
t = np.empty(n_lines-1)
E1 = np.empty(n_lines-1)
E2 = np.empty(n_lines-1)
E3 = np.empty(n_lines-1)
B1 = np.empty(n_lines-1)
B2 = np.empty(n_lines-1)
B3 = np.empty(n_lines-1)
kin = np.empty(n_lines-1)
mom1 = np.empty(n_lines-1)
mom2 = np.empty(n_lines-1)
mom3 = np.empty(n_lines-1)

file = open(filename, 'r')

line_num = 0
for line in file.readlines():
    if line_num > 0 :
        vals = line.rstrip().split(' ')
        t[line_num-1] = vals[0]
        E1[line_num-1] = vals[1]
        E2[line_num-1] = vals[2]
        E3[line_num-1] = vals[3]
        B1[line_num-1] = vals[4]
        B2[line_num-1] = vals[5]
        B3[line_num-1] = vals[6]
        kin[line_num-1] = vals[7]
        mom1[line_num-1] = vals[8]
        mom2[line_num-1] = vals[9]
        mom3[line_num-1] = vals[10]
    line_num = line_num + 1
```

Plot results (set variable fields to desired plottet quatity: E1, E2, E3, B1, B2, B3, kin, mom1, mom2, mom3, gauss):

```python
field = 'B3'

exec("plt.semilogy(t," + field + ", linewidth=1)")
plt.legend((field))
```
